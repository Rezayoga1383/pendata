# Tugas 3 Klasifikasi Naive Bayes Pada Datairis

**Langkah 1 : Mengimport Library**

langkah pertama yaitu kita mengimpor library numpy, pandas, dan matplotlib. 


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Langkah 2 Mengimpor Data Iris**

langkah selanjutnya kita memasukkan data iris. disini data iris saya import di GithubGist dengan nama file datairis.csv lalu kita panggil file tersebut ke dalam code. setelah itu kita akan menetapkan 4 variabel independen dari dataset ke variabel x dan variabel dependen 'species' ke variabel y. setelah itu kita akan menampilkan data 5 baris pertama dari data iris tersebut menggunakan fungsi head(). berikut adalah code dan outputnya.

```python
dataset = pd.read_csv('https://gist.githubusercontent.com/Rezayoga1383/2ccefe0ca9cb007642dbbbcf205c4336/raw/7dfd3f182984c41df3f6c0564efe235b9720cfd9/datairis.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

**Langkah 3  Memisahkan himpunan data ke dalam Kumpulan pelatihan dan Kumpulan pengujian**

Setelah kita memperoleh kumpulan data tadi, langkah selanjutnya kita harus membagi data menjadi set pelatihan dan set pengujian. Dalam kumpulan data ini, ada 150 baris dengan 50 baris dari masing-masing 3 kelas. Karena setiap kelas diberikan dalam urutan kontinu, kita perlu membagi himpunan data secara acak. Di sini, kami memiliki test_size=0.2, yang berarti bahwa 20% dari himpunan data akan digunakan untuk tujuan pengujian sebagai set pengujian dan 80% sisanya akan digunakan sebagai set pelatihan untuk melatih model klasifikasi Naive Bayes.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**Langkah 4 Fitur Scalling**

Himpunan data diperkecil ke rentang yang lebih kecil menggunakan opsi Penskalaan Fitur. Dalam hal ini, nilai X_train dan X_test diperkecil menjadi nilai yang lebih kecil untuk meningkatkan kecepatan program.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Langkah 5 Model Klasifikasi Naive Bayes pada Data Set**

Pada langkah ini, kami memperkenalkan kelas GaussianNB yang digunakan dari perpustakaan sklearn.naive_bayes. Di sini, kami telah menggunakan model Gaussian, ada beberapa model lain seperti Bernoulli, Categorical dan Multinomial. Di sini, kami menetapkan kelas GaussianNB ke pengklasifikasi variabel dan menyesuaikan nilai X_train dan y_train untuk tujuan pelatihan.

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**Langkah 6 Memprediksi hasil set Tes**

Setelah model dilatih, kita menggunakan classifier.predict() untuk memprediksi nilai untuk set Test dan nilai yang diprediksi disimpan ke variabel y_pred.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

**Langkah 7 Matriks Kebingungan dan Akurasi**

Ini adalah langkah yang banyak digunakan dalam teknik klasifikasi. Dalam hal ini, kita melihat Akurasi model terlatih dan memplot matriks kebingungan.

Matriks kebingungan adalah tabel yang digunakan untuk menunjukkan jumlah prediksi yang benar dan salah pada masalah klasifikasi ketika nilai sebenarnya dari Set Tes diketahui. Ini dari format

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

**Langkah 8 Membandingkan Nilai Riil dengan Nilai Prediksi**

Pada langkah ini, Pandas DataFrame dibuat untuk membandingkan nilai yang diklasifikasikan dari set Uji asli (y_test) dan hasil yang diprediksi (y_pred).

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```
