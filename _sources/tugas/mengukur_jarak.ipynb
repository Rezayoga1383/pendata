```python
import pandas as pd
from sklearn.metrics.pairwise import pairwise_distances
data = pd.read_csv('https://gist.githubusercontent.com/Rezayoga1383/2ccefe0ca9cb007642dbbbcf205c4336/raw/948d775ff62fe0e131d7a48dad54a4ef1008c9d3/data.csv')
data
```

```python
X = data[['Jenis Kelamin', 'Umur', 'Status', 'Mempunyai anak', 'Jabatan']].copy()
X['Jenis Kelamin'] = pd.Categorical(X['Jenis Kelamin']).codes
X['Umur'] = pd.Categorical(X['Umur']).codes
X['Status'] = pd.Categorical(X['Status']).codes
X['Mempunyai anak'] = pd.Categorical(X['Mempunyai anak']).codes
X['Jabatan'] = pd.Categorical(X['Jabatan'], categories=['CS', 'Satpam', 'manajer'], ordered=True).codes
```

```python
dist_matrix = pairwise_distances(X, metric='manhattan') / max(pairwise_distances(X, metric='manhattan').flatten())
```
