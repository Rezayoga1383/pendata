**1. Metode KNN**

Klasifikasi data KNN adalah salah satu aplikasi dari metode K-Nearest Neighbors (KNN) dalam pembelajaran mesin. Klasifikasi data KNN mengambil sekelompok data dengan fitur yang mirip dan mengklasifikasikannya ke dalam kelas atau label yang sama.

Dalam KNN, setiap data pada dasarnya diwakili sebagai sebuah titik dalam ruang fitur multidimensi. Ketika KNN melakukan klasifikasi data baru, ia akan mencari sekelompok tetangga terdekat dari data baru tersebut dalam ruang fitur tersebut. KNN kemudian mengambil mayoritas label dari tetangga terdekat sebagai label yang akan diberikan pada data baru tersebut. Oleh karena itu, KNN disebut sebagai metode "lazy learning" karena tidak memerlukan proses pelatihan yang kompleks seperti halnya dengan metode pembelajaran mesin yang lainnya.

**Langkah 1**

Import library yang diperlukan (pandas, numpy, sklearn)

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.preprocessing import LabelEncoder
```

**Langkah 2**

Import data yang ingin kita analisa

```python
df = pd.read_csv('https://gist.githubusercontent.com/Rezayoga1383/4a5ce7439abb0e49c53098c6fe61c263/raw/a2f25a0e53cf8c560cebd703e740ba7918a9e2e8/uts.csv')

table_row = df.shape[0]
df.head(table_row)
```

**Langkah 3**

Melakukan pengecekan apakah ada data yang kosong di dalam file data yang ingin kita analisa

```python
df_copy = df.copy()
df_copy.isnull().sum()
```

**Langkah 4**

Melakukan normalisasi data, di normalisasi data ini kita mengubah nilai yang ada di dalam fitur ke dalam bentuk bilangan biner atau rentan 0 hingga 1. 

Disini kita menggunakan metode Min-Max, di bawah ini adalah rumus dari metode Min-Max : 

$$
\text { Min }=(L \times R j)+S S \operatorname{Max}=2(L \times R j) \text { Min-Max }
$$

yang perlu digaris bawahi disini pada fitur target tidak akan dinormalisasi.

```python
target_column = df_copy['target']
df_copy = df_copy[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
df_copy = (df_copy - df_copy.min())/(df_copy.max() - df_copy.min())
df_copy['target'] = target_column.copy()
df_copy.head(table_row)
```

**Langkah 5**

1. Langkah pertama adalah menginisialisasi objek knn dari kelas KNeighborsRegressor, yang digunakan untuk melakukan regresi dengan algoritma K-Nearest Neighbors. Selanjutnya, kita melakukan randomisasi indeks dari dataset dengan menggunakan np.random.permutation() dan menetapkan seed 1 agar hasilnya konsisten.

2. Kemudian, dataset dibagi menjadi dua bagian: 75% untuk training dan 25% untuk testing. Data training disimpan dalam variabel train_index dan data testing disimpan dalam variabel testing_data.

3. Setelah itu, melakukan training model menggunakan data training dengan memanggil method fit() pada objek knn. Argumen input berupa fitur-fitur yang digunakan (yaitu age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, dan thal) dan labelnya adalah target.

4. Selanjutnya, model melakukan prediksi pada data testing dengan memanggil method predict() pada objek knn. Argumen input berupa fitur-fitur dari data testing yang ingin diprediksi.

5. Selanjutnya, akurasi model dihitung dengan mengukur berapa persen data testing yang berhasil diprediksi dengan threshold default 0.5.

6. Terakhir, hasil akurasi model akan dicetak.

```python
knn = KNeighborsRegressor()

# before we split, make sure you randomize the dataset
np.random.seed(1)
index_shuffled = np.random.permutation(df_copy.index)
shuffled_car_data = df_copy.loc[index_shuffled]

# We will use 75% of the data for training and the remaining 25% for testing
train_index = int(shuffled_car_data.shape[0] * 0.75)
training_data = shuffled_car_data.iloc[0:train_index]
testing_data = shuffled_car_data.iloc[train_index:]

# fit the model with the training data
knn.fit(training_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']], training_data['target'])

# predict the target of the testing data
prediction = knn.predict(testing_data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']])

# calculate the accuracy of the model
accuracy = np.mean(np.abs(prediction - testing_data['target']) <= 0.5)

print(f"Akurasi model KNN: {accuracy:.2%}")
```

**kesimpulan**

dengan hasil perhitungan di atas kita bisa mendapatkan nilai akurasi dimana dengan menggunakan metode KNN kita mendapatkan nilai akurasi sebesar 92.11%.

**Langkah 6**

Melakukan pemotongan data (Data Slicing)

1. pisahkan data menjadi 2 yaitu data training dan data test. disini kita menggunakan metode train_test_split dari sklearn.

2. Variabel features_train akan berisi data fitur untuk training, variabel features_test akan berisi data fitur untuk test, variabel target_train akan berisi target untuk training, dan variabel target_test akan berisi target untuk test.

```python
features = df_copy.values[:,:13]  #Variabel features diisi dengan nilai features dari dataset adult_df_rev dengan slicing [:,:13] artinya mengambil seluruh baris dan kolom 0 hingga kolom 3 (kolom ke-13 tidak diambil karena merupakan variabel target).
target = df_copy.values[:,13]  #Variabel target diisi dengan nilai target dari dataset adult_df_rev dengan slicing [:,13] artinya mengambil seluruh baris dan hanya kolom ke-13 (variabel target).
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size = 0.33, random_state = 10)  #Kemudian menggunakan fungsi train_test_split, dataset yang sudah dipisahkan dijadikan argumen, dengan tambahan argumen test_size = 0.33 yang berarti dataset akan dibagi menjadi 33% data uji dan 67% data latih, dan random_state = 10 sebagai seed untuk menghasilkan hasil yang sama setiap kali dijalankan.
```

**Langkah 7**

Menerapkan Gausian NaiveBayes dengan cara menggunakan modul GaussianNB dari sklearn.

```python
clf = GaussianNB()
clf.fit(features_train, target_train)
target_pred = clf.predict(features_test)
```

**Langkah 8**

Nilai akurasi model Naive Bayes

```python
accuracy = accuracy_score(target_test, target_pred, normalize = True)

print(f"Akurasi Model NaiveBayes: {accuracy:.2%}")
```
